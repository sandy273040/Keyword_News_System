```
import requests
from datetime import datetime

# 不同網站要改type代號
type = "B"

print("開始爬蟲")

# 修改爬蟲網址
r = requests.get(
    'https://cloud.feedly.com/v3/streams/contents?streamId=feed%2Fhttp%3A%2F%2Ftw.news.yahoo.com%2Frss%2Fbiz&ranked=engagement')
totalDict = r.json()
print("爬蟲結束")

# Get Item
totalItem = totalDict["items"]

# Connect to DB
import MySQLdb

conn = MySQLdb.connect(
    host='us-cdbr-east-02.cleardb.com',
    port=3306,
    user='bf6bdd17af0b08',
    passwd='0af9d14e',
    db='heroku_0f6b89cef5e8ad8'
)

# 建立連線
print("連線 DB")
cur = conn.cursor()

# For each news
print("Get News total rows=" + str(len(totalItem)))
i = 1
for newsDict in totalItem:
    print("Execute Row" + str(i))
    id = newsDict["id"]
    originId = newsDict["originId"]
    title = newsDict["title"]
    publishid = int(newsDict["published"])
    publishDate = datetime.fromtimestamp(int(str(publishid)[:-3])).strftime('%Y-%m-%d %H:%M:%S') #發布日期
    #publishDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    content = newsDict["summary"]["content"]
    transactiondate = datetime.now().strftime('%Y-%m-%d %H:%M:%S') #爬蟲當天日期

    # 塞資料前檢查塞過沒 by ID 檢查
    cur.execute("select * from news_data WHERE id = '{}' AND type = '{}'".format(id, type))  # 执行查询
    results = cur.fetchone()  # 拿到返回结果
    if results is None:
        print("id:" + id + " is not exists, INSERT Data:" + title)
        sql = "insert into news_data (id,orginid,type,title,publishid,publishdate,content,tranactiondate) " \
              "values('{}','{}','{}','{}',{},'{}','{}','{}')".format(id, originId, type, title, publishid, publishDate,
                                                                 content, transactiondate)
        print(sql)
        cur.execute(sql)
    else:
        print("id:" + id + " is exists, not INSERT Data:" + title)
    i += 1

# 關閉連線
cur.close()

# conn.commit()方法在提交事物，在向資料庫插入一條資料時必須要有這個方法，否則資料不會被真正的插入。
conn.commit()

# conn.close()關閉資料庫連線
conn.close()
```
